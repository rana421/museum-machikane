```python
# ベクトルデータベース
import openai
import os, json, re
from pprint import pprint

digit_pattern = re.compile(r'\d+')

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY")
openai.api_key = OPENAI_API_KEY

with open('./../classification/category.json', encoding="utf8") as f:
    classifications = json.load(f)

classification_numbers = [classification["category_number"] for classification in classifications]
classifications_dict = dict([classification.values() for classification in classifications])


index = []
for i, doc in enumerate(result):
    if doc[7] == "":
        class_text = "未分類"
    else:
        class_text = ",".join([classifications_dict[class_num] for class_num in digit_pattern.findall(doc[7]) if class_num in classification_numbers])
    text = f"{doc[1]}にある{doc[0]}で行われる{doc[2]}は{doc[3]}から{doc[4]}まで開催されます。その内容は{doc[6]}。分類：{class_text}"
    # ここでベクトル化を行う
    # openai.embeddings_utils.embeddings_utilsを使うともっとシンプルにかけます
    res = openai.Embedding.create(
        model='text-embedding-ada-002',
        input=text
    )

    # ベクトルをデータベースに追加
    index.append({
        'number': i+1,
        'name': doc[2],
        'raw': doc,
        'embedding_text': text,
        'embedding': res['data'][0]['embedding']
    })

with open('embedding_kansai.json', 'w', encoding="utf8") as f:
    json.dump(index, f)
```
